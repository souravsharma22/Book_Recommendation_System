```python
import pandas as pd 
import numpy as np 
```

```python
Users = pd.read_csv('./data/Users.csv')
Books = pd.read_csv('./data/Books.csv')
Ratings = pd.read_csv('./data/Ratings.csv')

ratings_with_name = Ratings.merge(Books, on='ISBN')
```

```python
ratings_with_name.head(2)
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
```

```python
reader_users = x[x].index
```

```python
filtered_rating  = ratings_with_name[ratings_with_name['User-ID'].isin(reader_users)]
```

```python
filtered_rating.shape
```

```python
filtered_rating.head(2)
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
famous_books = y[y].index
```

```python
famous_books.shape
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.head(2)
```

```python
table = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
table.head(2)
```

```python
table.fillna(0, inplace=True) 
```

```python
table.head(2)
```

```python
table.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(table)
```

```python
similarity_score.shape
```

```python
similarity_score[0].shape
```

```python
similarity_score
```

```python
table
```

```python
print(np.where(table.index =='Zoya')[0][0])
print(np.where(table.index =='A Bend in the Road')[0][0])
print(np.where(table.index =='Year of Wonders')[0][0])
```

```python
sorted(list(enumerate(similarity_score[0])),key= lambda x:x[1],reverse=True)[1:11]
```

```python
# creating a function to find this similar books 
def recommend(book_name):
    index = np.where(table.index ==book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])),key= lambda x:x[1],reverse=True)[1:6]

    for i in similar_books:
        print(table.index[i[0]])
    
```

```python
recommend('Zen and the Art of Motorcycle Maintenance: An Inquiry into Values')
```

```python
recommend('You Belong To Me')
```
